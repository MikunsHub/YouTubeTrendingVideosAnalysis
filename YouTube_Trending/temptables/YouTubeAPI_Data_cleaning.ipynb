```python
hours_pattern = re.compile(r"(\d+)H")
minutes_pattern = re.compile(r"(\d+)M")
seconds_pattern = re.compile(r"(\d+)S")


tot_hours = []
tot_minutes = []
tot_seconds = []
duration = []
# snippet_stat_content_join["duration"]
for i in snippet_stat_content_join["duration"]:
    hours = hours_pattern.search(i)
    minutes = minutes_pattern.search(i)
    seconds = seconds_pattern.search(i)

    hours = int(hours.group(1)) if hours else 0
    minutes = int(minutes.group(1)) if minutes else 0
    seconds = int(seconds.group(1)) if seconds else 0

    video_seconds = timedelta(
        hours = hours,
        minutes = minutes,
        seconds = seconds
    ).total_seconds()
    temp = video_seconds
    minutes,seconds = divmod(int(temp),60)
    hours,minutes = divmod(minutes,60)
    tot_hours.append(hours)
    tot_minutes.append(minutes)
    tot_seconds.append(seconds)

snippet_stat_content_join["hrs_duration"] = tot_hours
snippet_stat_content_join["min_duration"] = tot_minutes
snippet_stat_content_join["sec_duration"] = tot_seconds
```
